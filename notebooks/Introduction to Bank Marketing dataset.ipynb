## Modelling

```python
df.head()
```

```python
df['housing'].value_counts().sum
```

```python
df['default'] = df['default'].map({'yes':1, 'no':0})
df.head()
```

```python
df['housing'] = df['housing'].map({'yes':1, 'no':0})
df.head()
```

```python
df['loan'] = df['loan'].map({'yes':1, 'no':0})
df.tail()
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
# split the data into trainig and test set

x = df.drop(['job' , 'marital' , 'education' ,'contact' ,'month', 'poutcome' , 'deposit'] , axis = 1)
y = df['deposit']
```

```python
x.shape
```

```python
y.shape
```

```python
np.random.seed(0)
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
np.random.seed(0)


from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier


model1 = LogisticRegression()
model1.fit(x_train , y_train)
modell = model1.score(x_test , y_test)

model2 = KNeighborsClassifier()
model2.fit(x_train , y_train)
modelk = model2.score(x_test , y_test)

model3 = RandomForestClassifier()
model3.fit(x_train, y_train)
modelr = model3.score(x_test , y_test)


total_model = {
    'Logistic_Regression': modell*100,
    'KNeighbour_classifier' :modelk*100,
    'RandomForestClassifier': modelr*100
}
```

```python
total_model
```

```python
plt.style.use('seaborn')
model_compare = pd.DataFrame(data = total_model , index = ['accuracy'])
model_compare.T.plot(kind = 'bar')
plt.xticks(rotation =0)
```

# Hyperparameters tuning with Randomized Search CV

```python
# create hyperparameter for Logistic regression

log_reg_grid = {
    'C' : np.logspace(-4 , 4 , 20),
    'solver' : ['liblinear']
}

# create hyperparameters for Random forest classifier

rf_grid = {
    'n_estimators' : np.arange(10 , 1000 , 50),
    'max_depth' : [None , 3 , 5 , 10 ],
    
}
```

```python
# Tuning logistic regression
from sklearn.model_selection import RandomizedSearchCV , GridSearchCV
```

```python
np.random.seed(0)
rs_log_reg = RandomizedSearchCV(LogisticRegression() , param_distributions = log_reg_grid , n_iter = 30 , cv = 7 , verbose = True)

rs_log_reg.fit(x_train , y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(x_test , y_test)
```

```python
# tune for random forest classifier

rs_rf = RandomizedSearchCV(RandomForestClassifier() , param_distributions = rf_grid , n_iter = 35 , cv = 7 , verbose = True)

rs_rf.fit(x_train , y_train)
```

```python
rs_rf.best_params_
```

```python
rs_rf.score(x_test , y_test)
```

```python
# plot the tuned models

tuned_models = {
    'Logistic regression' : (rs_log_reg.score(x_test , y_test))*100,
    'Random forest classifier' : (rs_rf.score(x_test , y_test))*100
}

tuned_models
```

```python
# tuned model plot using dataframe

tuned_model_df = pd.DataFrame(data = tuned_models , index = ['accuracy'])
tuned_model_df.T.plot(kind = 'bar')
plt.xticks(rotation=0)
```
