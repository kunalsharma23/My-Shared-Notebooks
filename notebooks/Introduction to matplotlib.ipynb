## Plotting with pandas using OO method

```python
over_50=heart_disease[heart_disease['age']>50]
```

```python
over_50
```

```python
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',
            y='chol',
            c='target',ax=ax)
ax.set_xlim([45,100]);
```

```python
# Make a bit more of complicated plot

fig , ax =plt.subplots(figsize=(10,6))

scatter=ax.scatter(over_50['age'],
                  over_50['chol'],
                  c=over_50['target'])

ax.set(title='Heart disease and cholrstrol Level', xlabel='age', ylabel='cholesterol')
ax.legend(*scatter.legend_elements(),title='target');
```

```python
# Make a bit more of a complicated plot by adding a mean line

fig , ax=plt.subplots(figsize=(10,6))

scatter=ax.scatter(over_50['age'],
                  over_50['chol'],
                  c=over_50['target'])

ax.set(title='Heart disease and cholesterol LeveL', xlabel='age 19 munde di',ylabel='Cholestrol')

ax.legend(*scatter.legend_elements(),title='target')

ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
## Adding another plot to the existing styled one

fig , (ax0,ax1)=plt.subplots(nrows=2, ncols=1,sharex=True,figsize=(10,8))

scatter=ax0.scatter(over_50['age'],over_50['chol'],c=over_50['target'])

ax0.set(title='Heart disease and cholesterol level',ylabel='cholestrol',)

ax0.legend(*scatter.legend_elements(),title='targets')

ax0.axhline(y=over_50['chol'].mean(), color='b',linestyle='--',label='Average')



scatter=ax1.scatter(over_50['age'],over_50['thalach'],c=over_50['target'])

ax1.set(title='Heart disease and heart rate',xlabel='age',ylabel='Maximum heart rate')

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(y=over_50['thalach'].mean(),color='b',linestyle='--',label='Average');



fig.suptitle('Heart Disease Analysis',fontsize=18,fontweight='bold')
```

## changing the cmap

```python
over_50
```

```python


fig , ax = plt.subplots(figsize=(10,6))

scatter=ax.scatter(over_50['age'], over_50['chol'],over_50['target'])

ax.set(title='Heart disease and cholestrol level', xlabel='Age', ylabel='Cholestrol')

ax.axhline(y=over_50['chol'].mean(),c='b', linestyle='--',label='Average')

ax.legend(*scatter.legend_elements(),title='Target');
```

```python
# No cmap change
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], cmap='winter')
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.axhline(y=over_50["chol"].mean(), 
           c='r', 
           linestyle='--', 
           label="Average");
ax.legend(*scatter.legend_elements(), title="Target");
```

```python
## Adding another plot to the existing styled one

fig , (ax0,ax1)=plt.subplots(nrows=2, ncols=1,sharex=True,figsize=(10,8))

scatter=ax0.scatter(over_50['age'],over_50['chol'],c=over_50['target'],cmap='winter')

ax0.set(title='Heart disease and cholesterol level',ylabel='cholestrol',)

ax0.legend(*scatter.legend_elements(),title='targets')

ax0.axhline(y=over_50['chol'].mean(), color='b',linestyle='--',label='Average')



scatter=ax1.scatter(over_50['age'],over_50['thalach'],c=over_50['target'],cmap='summer')

ax1.set(title='Heart disease and heart rate',xlabel='age',ylabel='Maximum heart rate')

ax1.legend(*scatter.legend_elements(),title='Target')

ax1.axhline(y=over_50['thalach'].mean(),color='b',linestyle='--',label='Average');



fig.suptitle('Heart Disease Analysis',fontsize=18,fontweight='bold')
```
